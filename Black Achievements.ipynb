### Research Questions

1. How many Black women are featured in the technological innovations?
2. Have the achievements gone up or down over time?


#### The variables I would need to answer these questions are as follows:

1. "gender" column which designates whether a woman achieved each accomplishment or not though the data does not seem to make room for non-binary people
2. "year" column will denote how many accomplishments have occured in each year and will help reveal trends about how accomplishments have increased or decreased over time

```python
# import libraries for viewing and analyzing data

import pandas as pd
```

```python
# this designation does not seem as though it will be as definitive and thorough as I would like
# however, I do not want to go through and assign gender based on names

# now to see the unique values in the "category" column
firstsDF['category'].unique()
```

```python
# I can search the actual accomplishments for the words "woman" or "female" using a for loop and a counter

counter = 0

for i in range (len(firstsDF.accomplishment)):
    if "woman" in firstsDF.accomplishment[i] or "female" in firstsDF.accomplishment[i]:
        print (firstsDF.accomplishment[i])
        counter += 1
print(counter)
```

```python
# now I want to create a new dataFrame with accomplishments including "woman" and "female"

firstsWomen = firstsDF.loc[firstsDF['accomplishment'].str.contains("woman|female", case=False)]

firstsWomen
```

```python
# I can also check the gender column to confirm how many "Female" entries there are

# first I will use a for loop

counter = 0

for i in range (len(firstsDF.gender)):
    if "Female" in firstsDF.gender[i]:
        print(firstsDF.gender[i])
        counter += 1
        
print(counter)

```

```python
# interesting!  So there are some accomplishments attributed to women that do not include the words "woman" or "female"

# now I will create a new dataFrame from the gender column where the word "Female" is present

firstsFemale = firstsDF.loc[firstsDF['gender'].str.contains("Female", case=False)]

firstsFemale
```

```python
# just to close the loop on this, I want to find the accomplishments that do not include the words "woman" or "female" 
# but who have "Female" in the gender

firstsFemaleNoGender = firstsFemale.loc[~firstsFemale['accomplishment'].str.contains("female|woman", case=False)]

# here are the entries I would have missed if I just went with accomplishments that had the word "woman" or "female"
# looking at the list, I am not sure why some of the entries have "Female" designations, so perhaps looking at gender may not 
# be as fruitful as I had hoped it would be

firstsFemaleNoGender
```

```python
# I would like to plot a bar graph that shows the breakdown of accomplishments by gender present in the data

# Calculate the values for the x and y axis
x = list(firstsDF['gender'].unique())
y = firstsDF['gender'].value_counts()

# Create a express bar chart 
fig = px.bar(x=x, y=y)

# Style the figure 
fig.update_layout(title='African American Achievements by Gender',
                  xaxis_title='Gender', 
                  yaxis_title='Accomplishments')
fig.show()
```

```python
# I would like to plot a bar graph that shows the breakdown of accomplishments across time

# Calculate the values for the x and y axis
x = list(firstsDF['year'].unique())
y = firstsDF['year'].value_counts()

# Create a express bar chart 
fig2 = px.bar(x=x, y=y)

# Style the figure 
fig2.update_layout(title='African American Achievements by Year',
                  xaxis_title='Year', 
                  yaxis_title='Accomplishments')
fig2.show()

# the accomplishments seem to be on a declining trajectory over time
```
